```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline# Import the image
img=cv2.imread("burano.png")
plt.imshow(img)
```

```python
#convert the image into RGB
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
```

```python
#plot the three channels of the image,RGb is made up of stacking these three channels
fig,axs=plt.subplots(nrows=1,ncols=3,figsize=(20,20))
for i in range(0,3):
    ax=axs[i]
    ax.imshow(img_rgb[:,:,i],cmap='gray')
plt.show()
```

```python
#transform the image into HSV and HLS models
img_hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
img_hls=cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
#plot images
fig,(ax1,ax2)= plt.subplots(nrows=1,ncols=2,figsize=(20,20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
```

```python
#Drawing on image
img_2=cv2.imread('img2.png')
img_copy=img_2.copy()
#draw a rectangle
cv2.rectangle(img_copy,pt1=(800,470),pt2=(980,530),color=(255,0,0),thickness=5)
plt.imshow(img_copy)
```

```python
cv2.circle(img_copy,center=(950,50),radius=50,color=(0,0,255),thickness=5)
plt.imshow(img_copy)
```

```python
#add text
cv2.putText(img_copy,text="The wall of love",org=(250,250),
           fontFace=cv2.FONT_HERSHEY_DUPLEX,fontScale=2,color=(0,255,0),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(img_copy)
```

```python
#Draw circles on mouse button
#Step1: define callback function
def draw_circle(event,x,y,flags,param):
    if event==cv2.EVENT_LBUTTONUP:
        cv2.circle(img_copy,center=(x,y),radius=5,color=(87,184,237),thickness=-1)
    elif event==cv2.EVENT_BUTTONDOWN:
        cv2.circle(img,center=(x,y),radius=10,color=(87,184,237),thickness=1)

#call the window

img=cv2.imread('map.png')
cv2.namedWindow(winname="my_drawing")
cv2.setMouseCallback('my_drawing',draw_circle)

#step3 execution
while True:
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(10)& 0XFF==27:
        break

cv2.destroyAllWindows()
```
